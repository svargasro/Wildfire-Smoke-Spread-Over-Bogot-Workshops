```python
import pandas as pd
import numpy as np
```

```python
estaciones = ['centro de alto rendimiento', 'guaymaral', 'las ferias', 'minambiente', 'suba', 'usaquen', 'ciudad bolivar', 
                 'colina', 'jazmin', 'movil fontibon', 'usme', 'san cristobal', 'tunal', 'bolivia', 'carvajal - sevillana', 
                 'fontibon', 'kennedy', 'movil 7ma', 'puente aranda']


cont_base = pd.read_csv('../Data/bolivia_2024-01-22_31.csv')['Fecha Hora'].to_frame()
```

```python
for estacion in estaciones:
    df2023 = pd.read_csv(f'../Data/{estacion}_2023-01-23_2023-02-01.csv')
    df2024 = pd.read_csv(f'../Data/{estacion}_2024-01-22_31.csv')

    df2023[['PM10 [µg/m3]', 'PM2.5 [µg/m3]']] = df2023[['PM10 [µg/m3]', 'PM2.5 [µg/m3]']].apply(lambda x: x.astype(str).str.replace(',', '.').astype(float), axis=0)
    df2024[['PM10 [µg/m3]', 'PM2.5 [µg/m3]']] = df2024[['PM10 [µg/m3]', 'PM2.5 [µg/m3]']].apply(lambda x: x.astype(str).str.replace(',', '.').astype(float), axis=0)

    #diff_PM10 = np.array(df2024['PM10 [µg/m3]'] - df2023['PM10 [µg/m3]'])
    diff_PM2_5 = np.array(df2024['PM2.5 [µg/m3]'] - df2023['PM2.5 [µg/m3]'])

    #diff_PM10 = np.where(diff_PM10 < 0, 0, diff_PM10)
    diff_PM2_5 = np.where(diff_PM2_5 < 0, 0, diff_PM2_5)

    #cont_base[f'{estacion}_PM10 [µg/m3]'] = diff_PM10
    cont_base[f'{estacion}_PM2.5 [µg/m3]'] = diff_PM2_5
```
