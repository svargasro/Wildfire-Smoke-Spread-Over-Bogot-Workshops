```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos de los archivos de texto

#theta
pendulos_theta = np.loadtxt('PendulosTheta.txt')
cuna_theta = np.loadtxt('CunaNewtonTheta.txt')

#omega
pendulos_omega = np.loadtxt('PendulosOmega.txt')
cuna_omega = np.loadtxt('CunaNewtonOmega.txt')

#Tau
pendulos_alpha = np.loadtxt('PendulosAlpha.txt')
cuna_alpha = np.loadtxt('CunaNewtonAlpha.txt')

cuna_theta.shape, cuna_omega.shape, cuna_alpha.shape
```

```python
# Selecciona cada 4 elementos de un arreglo
# siendo la sintaxis [inicio:fin:paso]
# y se guarda en un nuevo arreglo

pendulos_theta = pendulos_theta[::5]
pendulos_omega = pendulos_omega[::5]
pendulos_alpha = pendulos_alpha[::4]

cuna_theta = cuna_theta[::5]
cuna_omega = cuna_omega[::5]
cuna_alpha = cuna_alpha[::4]


t=0
dt=1e0
tmax=100
time = np.arange(t,tmax,dt)

time.shape, cuna_alpha.shape
```

```python
#@title Graficar de Theta

# Graficar los datos
plt.title('Theta vs Tiempo')
plt.plot(time, pendulos_theta,"-.r", label='Pendulo')
# plt.plot(time, cuna_theta,"-.b", label='CunaNewton')
plt.xlabel('Tiempo')
plt.ylabel('Theta')
plt.legend()

# Realizar una estimacin multilnea
# Aqu puedes agregar tu cdigo para realizar la estimacin multilnea

# Mostrar la grfica
plt.show()
```
