```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# Directorio donde están los archivos
directory = os.getcwd()

# Lista para guardar los DataFrames
data_frames = []

# Lista de colores para diferenciar cada gráfico
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']
color_index = 0
```

```python
# Configuración inicial de la gráfica
plt.figure(figsize=(10, 6))

# Leer todos los archivos en el directorio
for file in os.listdir(directory):
    if file.startswith("C") and file.endswith(".csv"):
        # Construir el path completo del archivo
        file_path = os.path.join(directory, file)
        # Leer el archivo asumiendo punto y coma como delimitador y sin encabezado
        data = pd.read_csv(file_path, delimiter=';', header=None, names=['Time', 'Value'])
        # Añadir el DataFrame a la lista
        data_frames.append(data)
        # Graficar
        plt.plot(data['Time'], data['Value'], label=file[:-4], color=colors[color_index])
        color_index = (color_index + 1) % len(colors)

# Añadir leyenda, título y etiquetas
plt.legend()
plt.title('Comparative Plot of CSV Files')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
